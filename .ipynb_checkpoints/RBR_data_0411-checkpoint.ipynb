```python
def multiply_minutes_and_seconds(input_time):
    # Split the input time into hours, minutes, and seconds
    hours, minutes, seconds = map(int, input_time.split(':'))

    # Multiply minutes by 960
    mins = minutes * 960
    secs = seconds * 16
    
    return mins+secs

def subtract_times(time1, time2):
    # Convert input times to datetime.time objects
    time1_obj = datetime.strptime(time1, '%H:%M:%S').time()
    time2_obj = datetime.strptime(time2, '%H:%M:%S').time()

    # Subtract the times to get the time difference
    time_diff = datetime.combine(datetime.min, time1_obj) - datetime.combine(datetime.min, time2_obj)

    # Extract the time difference as a string in the "HH:MM:SS" format
    time_diff_str = str(time_diff).split(':')
    hours = int(time_diff_str[0])
    minutes = int(time_diff_str[1])
    seconds = int(time_diff_str[2])

    # Return the time difference as a tuple of hours, minutes, and seconds
    return "{:02d}:{:02d}:{:02d}".format(hours, minutes, seconds)

#making marker plot function to plot each marker we need
def thermistors_oneminute_plot(start_time):
    for i in range(len(divers_oneminute_timeseries)):
        time_difference1 = subtract_times(divers_oneminute_timeseries.index[i], start_time)
        time_difference2 = subtract_times(divers_oneminute_timeseries.end_time[i], start_time)

        marker_timevalues = [multiply_minutes_and_seconds(time_difference1), multiply_minutes_and_seconds(time_difference2)]
        #marker_tempvalues = [T2_timeseries.loc[times_oneminute_timeseries.index[i], "Temperature"].values[0], T2_timeseries.loc[times_oneminute_timeseries.index[i], "Temperature"].values[0]]
        #starts at minimum
        marker_tempvalues = [23.5, 23.5]
        colors = ["red", "blue", "lightcoral", "darkblue", "hotpink", "plum", "orchid", "seagreen", "purple", "peru", "skyblue", "green"]
        plt.scatter(marker_timevalues, marker_tempvalues, color=colors[i], marker='|', s=500000, label=i+1)
        

def thermistors_original_plot(start_time):
    for i in range(len(divers_original_timeseries)):
        time_difference1 = subtract_times(divers_original_timeseries.index[i], start_time)
        time_difference2 = subtract_times(divers_original_timeseries.end_time[i], start_time)

        marker_timevalues = [multiply_minutes_and_seconds(time_difference1), multiply_minutes_and_seconds(time_difference2)]
        #marker_tempvalues = [T2_timeseries.loc[times_oneminute_timeseries.index[i], "Temperature"].values[0], T2_timeseries.loc[times_oneminute_timeseries.index[i], "Temperature"].values[0]]
        #starts at minimum
        marker_tempvalues = [23.5, 23.5]
        colors = ["red", "blue", "lightcoral", "darkblue", "hotpink", "plum", "orchid", "seagreen", "purple", "peru", "skyblue", "green"]
        plt.scatter(marker_timevalues, marker_tempvalues, color=colors[i], marker='|', s=500000, label=i+1)

#zorder=3
```
